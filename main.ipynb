```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("train.csv")
testdata = pd.read_csv("test.csv")
dataset.head()
```

```python
dataset = dataset.drop(["Unnamed: 0", "id"], axis=1)
testdata = testdata.drop(["Unnamed: 0", "id"], axis=1)
```

```python
dataset = dataset.replace("neutral or dissatisfied", value=0, regex=True)
dataset = dataset.replace("satisfied", value=1, regex=True)
dataset["satisfaction"] = pd.to_numeric(dataset["satisfaction"]) #to cahnge datatype
```

```python
testdata = testdata.replace("neutral or dissatisfied", value=0, regex=True)
testdata = testdata.replace("satisfied", value=1, regex=True)
testdata["satisfaction"] = pd.to_numeric(testdata["satisfaction"])
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X_test = testdata.iloc[:, :-1].values
y_test = testdata.iloc[:, -1].values
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy="most_frequent")
si.fit(X[:, 21:22])
si.fit(X_test[:, 21:22])
X[:, 21:22] = si.transform(X[:, 21:22])
X_test[:, 21:22] = si.transform(X_test[:, 21:22])
```

```python
np.isnan(np.min(X[:, 21:22])) 
```

```python
dataset.boxplot(["Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"])
```

```python
li = ["Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]
maxt = []
mint = []
for i in li:
    maxt.append(dataset[i].quantile(0.95))
    mint.append(dataset[i].quantile(0.05))
```

```python
print(maxt)
print(mint)
```

```python
a, b = 0, 0
for i in li:
    
    dataset = dataset[(dataset[i]<maxt[a]) & (dataset[i]>mint[b])]
    testdata = testdata[(testdata[i]<maxt[a]) & (testdata[i]>mint[b])]
    a+=1
    b+=1
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0, 1, 3, 4])], remainder = "passthrough")
X = np.array(ct.fit_transform(X))
X_test = np.array(ct.fit_transform(X_test))
```

```python
X.shape
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier()
xg.fit(X, y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = xg.predict(X_test)
c = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
ac = cross_val_score(estimator=xg, X=X, y=y, cv = 10)
print(ac.mean()*100)
```

```python
print(c)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X, y)
y_predrf = rf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_predrf))
```

```python
from catboost import CatBoostClassifier
cb = CatBoostClassifier()
cb.fit(X, y)
y_predcb = cb.predict(X_test)
```

```python
print(accuracy_score(y_test, y_predcb))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X, y)
y_predk = knn.predict(X_test) 
```

```python
print(accuracy_score(y_test, y_predk))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X, y)
y_prednb = nb.predict(X_test)
```

```python
print(accuracy_score(y_test, y_prednb))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X, y)
y_preddt = dt.predict(X_test)
```

```python
print(accuracy_score(y_test, y_preddt))
```
